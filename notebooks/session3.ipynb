## Part 1: Data Accessibility Constraints

### Background: The Original Exiobase Distribution

Exiobase was originally published following academic data sharing conventions:
- **Repository**: Zenodo (open science platform, DOI-based versioning)
- **Format**: ZIP archives containing CSV matrices and metadata files
- **Structure**: Separate files for monetary flows, environmental extensions, concordance tables
- **Documentation**: Academic papers, technical reports, scattered README files
- **Access pattern**: Download entire archives (hundreds of MB), extract locally, reconstruct table relationships

This distribution model works well for researchers conducting deep analyses who will spend weeks with the data. It creates friction for exploratory work or automated pipelines.

### Cognitive and Technical Barriers

**For human analysts:**
- Understanding MRIO structure requires methodological background
- Multiple files must be combined correctly (matrix multiplication, proper indexing)
- Classification systems (ISIC, NACE, CPA) require domain knowledge
- Data cleaning and reshaping before analysis can begin

**For coding agents:**
- Zenodo requires authentication or interactive download (no direct URL to data)
- ZIP extraction and file navigation with unclear naming conventions
- Schema inference from academic documentation rather than formal standards
- Limited training data representation (specialized domain)

## Part 3: Methodological Reconciliation

### Background: Why Numbers Don't Match

OWID and Exiobase measure related but distinct concepts:

**OWID (Our World in Data)**:
- Source: Primarily CDIAC, GCP, UNFCCC national inventories
- Scope: **Territorial emissions** (produced within borders)
- Coverage: COâ‚‚ from fossil fuels and industry, land use separately
- Reporting: Annual national totals with standardized time series

**Exiobase**:
- Source: MRIO model combining national accounts, trade data, emission factors
- Scope: **Consumption-based emissions** (allocated to final consumers) or production-based depending on table
- Coverage: Multi-gas (COâ‚‚, CHâ‚„, Nâ‚‚O) with sectoral attribution
- Reporting: Input-output tables requiring post-processing for national totals

**Additional discrepancies:**
- Year of data publication (OWID more current, Exiobase has processing lag)
- Emission factor assumptions (fuel carbon content, combustion efficiency)
- Sector boundary definitions (is international aviation included? bunker fuels?)
- Allocation of multi-region activities (electricity imports, trade embodied emissions)

**The analytical challenge**: When two authoritative sources disagree, you must understand methodological differences to interpret the gap, not assume one is "correct."
